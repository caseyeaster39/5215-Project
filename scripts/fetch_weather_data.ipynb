```python
shuffled_df = df.sample(frac = 1)
start_time = time.time()
begin_checkpoint = time.time()
num_nulls = 0
row_count = 0
for index, row in shuffled_df.iterrows():
    check_temp = row['temp']
    if not pd.isnull(check_temp):
        num_nulls += 1
        continue

    lat = row['Latitude']
    lng = row['Longitude']
    hour = row['Hour']
    month = row['Month']
    #print("row", row.to_string())

    day = random.randint(2, 27)
    start = datetime(2019, month, day)
    end = datetime(2019, month, day + 1)
    location = Point(lat, lng, 70)

    #print("Hour:", hour)
    data = Hourly(location, start, end)
    data = data.fetch()
    hour_row = data.iloc[hour,:].to_dict()
    for item in hour_row.keys():
        #print(item, "is type", type(item), ":", hour_row[item], "is type", type(hour_row[item]))
        shuffled_df.loc[index,item] = hour_row[item]
    #print("keys", hour_row.keys())
    #print(hour_row)
    # Plot line chart including average, minimum and maximum temperature
    # hour_row.plot(y=['tavg', 'tmin', 'tmax'])
    # plt.show()
    # First 1000 took ~28sec
    # 2000 skipping first 1000 took 30.80380606651306 seconds
    # 3000 skipping first 2000 took 27.952234029769897 seconds
    # 9000 skipping 6000 rows took 79.95096492767334 seconds
    # 6000 skipping 3000 rows took 85.12703514099121 seconds
    if row_count != 0 and (row_count + 1) % 5 ==  0:
        print("--------------------------------------")
        checkpoint_time = time.time()
        csv_name = "../ref/train_with_zip_pop_weather" + str((row_count + 1)) + ".csv"

        print("This iteration has taken", (checkpoint_time - begin_checkpoint), "seconds")
        begin_checkpoint = time.time()

        print("Run has taken", (checkpoint_time - start_time), "seconds")

        print("Completed", (row_count + 1), "rows. Saving to csv:", csv_name)
        print("Saved to:", csv_name)
        shuffled_df.to_csv(csv_name, index=False)
        print("Skipped", num_nulls, "rows this iteration")
        print("--------------------------------------\n")
    row_count += 1
```
